# Output Evaluation

## SPC

```python
import json
from collections import Counter
import matplotlib.pyplot as plt
```

### Consistency

```python
file_path = 'score_consistency.json'

with open(file_path, 'r') as file:
    data = json.load(file)

cleaned_data = [item.strip() for item in data]

counts = Counter(cleaned_data)

print(counts)
```

```python
total = sum(counts.values())
percent_1 = (counts['-1'] / total) * 100
percent_2 = (counts['0'] / total) * 100
percent_3 = (counts['1'] / total) * 100
```

```python
labels = ['Inconsistency', 'Neural', 'Consistency']
percentages = [percent_1, percent_2, percent_3]

print(f"Percentage of '{labels[0]}'s: {percent_1:.2f}%\n")
print(f"Percentage of '{labels[1]}'s: {percent_2:.2f}%\n")
print(f"Percentage of '{labels[2]}'s: {percent_3:.2f}%\n")
```

```python
plt.figure(figsize=(10, 6))
plt.bar(labels, percentages, color=['blue', 'green', 'red', 'purple'])
plt.xlabel('Response Categories')
plt.ylabel('Percentage')
plt.title('Survey Responses: Consistency Rating')
plt.ylim(0, 100)
plt.show()
```
